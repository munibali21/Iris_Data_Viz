```python
import seaborn as sns
import matplotlib.pyplot as plt
flar=sns.load_dataset("iris")
flar
```

```python
sns.lineplot(x="sepal_length",y="sepal_width",data=flar)
plt.title("graph practice")
sns.set_style("dark")
sns.set_style(style=None,rc=None)
plt.show()
```

```python
sns.barplot(x="species",y="petal_length",data=flar)
plt.figure(figsize=(2,4))
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
ship=sns.load_dataset("titanic")
ship
```

```python
sns.barplot(x="sex",y="age",hue="who",data=ship,order=["male","female"])
plt.show
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
ship=sns.load_dataset("titanic")
sns.barplot(x="class",y="fare",data=ship,linewidth=3,facecolor=(1,1,1,1),errcolor="0",edgecolor="0")
```

```python
import seaborn as sns
res=sns.load_dataset("tips")
res
sns.boxplot(x="day",y="tip",hue="sex",data=res,palette="Set2",dodge=False)
```

```python
import seaborn as sns
ship=sns.load_dataset("titanic")
ship
sns.boxplot(x="survived",y="age"
            ,showmeans=True,
            meanprops={"marker":"*","markersize":"12","markeredgecolor":"green"}
            ,data=ship)
plt.show()
```
